```python
def decrypt_salary(enc_salary: str):
    f = Fernet(os.getenv("FERNET_KEY"))
    token_bytes = enc_salary.encode("utf-8")
    decrypted_bytes = f.decrypt(token_bytes)
    decrypted_str = decrypted_bytes.decode("utf-8")

    return int(decrypted_str)
```
